# Model Creation

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, f1_score, roc_auc_score, classification_report
from sklearn.ensemble import RandomForestClassifier
import xgboost as xgb
from xgboost import XGBClassifier
from collections import Counter
```

```python
df = pd.read_csv("../data/player_churn_data.csv")
df.head()
```

```python
features = [
    "total_playtime_hours",
    "sessions_last_30d",
    "sessions_last_7d",
    "purchases_count",
    "max_level",
    "friends_count"
]

X = df[features]
y = df["churned"]
```

## Train Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.20,
    random_state=42,
    stratify=y
)
```

```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
model = LogisticRegression()
model.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_test_scaled)
y_pred_prob = model.predict_proba(X_test_scaled)[:, 1]
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("F1 Score:", f1_score(y_test, y_pred))
print("ROC-AUC:", roc_auc_score(y_test, y_pred_prob))

print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

```python
coef = model.coef_[0]

coef_df = pd.DataFrame({
    "feature": features,
    "coefficient": coef
}).sort_values("coefficient", ascending=True)

plt.figure(figsize=(8,5))

colors = ["steelblue" if c > 0 else "salmon" for c in coef_df["coefficient"]]

plt.barh(coef_df["feature"], coef_df["coefficient"], color=colors)

plt.title("Logistic Regression Coefficients", fontsize=16, weight="bold")
plt.xlabel("Coefficient Value", fontsize=12)
plt.ylabel("Feature", fontsize=12)

for i, (coef_value) in enumerate(coef_df["coefficient"]):
    plt.text(
        coef_value * 1.05,
        i,
        f"{coef_value:.3f}",
        va="center",
        fontsize=10
    )

plt.grid(axis="x", linestyle="--", alpha=0.4)

plt.savefig("../plots/log_reg_coeffs.png", dpi=300, bbox_inches="tight")
plt.show()
```

## Random Forest

```python
rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=5,
    min_samples_split=20,
    min_samples_leaf=10,
    class_weight="balanced",
    random_state=42
)
```

```python
rf.fit(X_train, y_train)
rf_pred = rf.predict(X_test)
rf_pred_prob = rf.predict_proba(X_test)[:, 1]
```

```python
print("RF Accuracy:", accuracy_score(y_test, rf_pred))
print("RF F1 Score:", f1_score(y_test, rf_pred))
print("RF ROC-AUC:", roc_auc_score(y_test, rf_pred_prob))
print("\nClassification Report:\n", classification_report(y_test, rf_pred))
```

```python
rf_importances = rf.feature_importances_
indices = np.argsort(rf_importances)

sorted_features = np.array(features)[indices]
sorted_importances = rf_importances[indices]

plt.figure(figsize=(8,5))

colors = ["steelblue"] * len(sorted_importances)

plt.barh(sorted_features, sorted_importances, color=colors)

plt.title("Random Forest Feature Importance", fontsize=16, weight="bold")
plt.xlabel("Importance Score", fontsize=12)
plt.ylabel("Feature", fontsize=12)

for i, val in enumerate(sorted_importances):
    plt.text(
        val * 1.02,
        i,
        f"{val:.3f}",
        va="center",
        ha="left",
        fontsize=10
    )

plt.grid(axis="x", linestyle="--", alpha=0.4)

plt.xlim(0, sorted_importances.max() * 1.15)

plt.savefig("../plots/rf_feature_importance.png", dpi=300, bbox_inches="tight")
plt.show()
```

## XGBoost

```python
counter = Counter(y_train)
neg, pos = counter[0], counter[1]
scale_pos_weight = neg / pos

print("Class counts:", counter)
print("scale_pos_weight:", scale_pos_weight)
```

```python
xgb_model = XGBClassifier(
    n_estimators=300,
    learning_rate=0.1,
    max_depth=4,
    subsample=0.8,
    colsample_bytree=0.8,
    scale_pos_weight=scale_pos_weight,
    eval_metric="logloss",
    random_state=42,
    n_jobs=-1
)

xgb_model.fit(X_train, y_train)
```

```python
xgb_pred = xgb_model.predict(X_test)
xgb_pred_prob = xgb_model.predict_proba(X_test)[:, 1]

print("XGB Accuracy:", accuracy_score(y_test, xgb_pred))
print("XGB F1 Score:", f1_score(y_test, xgb_pred))
print("XGB ROC-AUC:", roc_auc_score(y_test, xgb_pred_prob))

print("\nClassification Report:\n", classification_report(y_test, xgb_pred))
```

```python
import numpy as np
import matplotlib.pyplot as plt

xgb_importances = xgb_model.feature_importances_
indices = np.argsort(xgb_importances)

sorted_features = np.array(features)[indices]
sorted_importances = xgb_importances[indices]

plt.figure(figsize=(8,5))

colors = ["steelblue"] * len(sorted_importances)

plt.barh(sorted_features, sorted_importances, color=colors)

plt.title("XGBoost Feature Importance", fontsize=16, weight="bold")
plt.xlabel("Importance Score", fontsize=12)
plt.ylabel("Feature", fontsize=12)

for i, val in enumerate(sorted_importances):
    plt.text(
        val * 1.02,
        i,
        f"{val:.3f}",
        va="center",
        ha="left",
        fontsize=10
    )

plt.grid(axis="x", linestyle="--", alpha=0.4)

plt.xlim(0, sorted_importances.max() * 1.15)

plt.savefig("../plots/xgb_feature_importance.png", dpi=300, bbox_inches="tight")
plt.show()
```
